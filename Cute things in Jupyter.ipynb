```julia
function hilb(T,n)
    H = zeros(T,n,n)
    for i = 1:n, j = 1:n
        H[i,j] = 1/(i+j)
    end
    return H # return as a tuple if multiple output
end

H = hilb(Complex64, 10)
```

```julia
function hilb(T,n)
    H = zeros(T,n,n)
    for i = 1:n, j = 1:n
        H[i,j] = 1/(i+j)
    end
    return H # return as a tuple if multiple output
end

#Multiple dispatch
#{T} 
# :: works on this particular type
function hilb{T}(::Type{Rational{T}},n)   #hilb{T<:Rational}(T,n)  #H = zeros(T,n,n)
    H = zeros(Rational{T},n,n)
    for i = 1:n, j = 1:n
        H[i,j] = 1/(i+j)
    end
    return H # return as a tuple if multiple output
end

```
